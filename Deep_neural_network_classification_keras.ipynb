```python
#Splitting the dataset between independent and dependent variables 
X = data.iloc[:,:-1]
Y = data.iloc[:,-1]
```

Building the model

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
cols = X.shape[1]

#Adding the layers
model.add(Dense(250, activation='relu', input_shape=(cols,)))
model.add(Dense(250, activation='relu'))
model.add(Dense(250, activation='relu'))
model.add(Dense(2, activation='softmax')) # The last layer contains Only 2 nodes , the patient has diabetes or not 
```

Compiling the model - 
A lowest score in categorical_crossentropy indicates the model performs better 

```python
import tensorflow as tf
model.compile(optimizer='adam',loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

Training the model

```python
from keras.callbacks import EarlyStopping

early_stopping_monitor = EarlyStopping(patience = 3)

model.fit(X,Y, epochs = 30, validation_split = 0.2, 
          callbacks=[early_stopping_monitor])
```
