```python
#Converting all null value columns to mean 
data[['T','TM','Tm','SLP','H','VV','V','VM','PM 2.5']] = data[['T','TM','Tm','SLP','H','VV','V','VM','PM 2.5']].fillna(data[['T','TM','Tm','SLP','H','VV','V','VM','PM 2.5']].mean())
```

```python
#Dividing the dataset to independent and dependent variables 
X = data.iloc[:,:-1]
Y = data.iloc[:,-1]
```

```python
from tensorflow import keras
from tensorflow.keras import layers
from kerastuner.tuners import RandomSearch
```

Building the model 

```python
def build_model(hp): #hp is the parameter , assigning the instance of the specific model 
  model = keras.Sequential() #adding the sequential layer 
  for i in range(hp.Int('num_layers',2,20)): #num of hidden layers will be between 2 and 20 , Int measn we will try with diff values 
    model.add(layers.Dense(units = hp.Int( 'units', #inside each hidden layer nodes are being created using dense function 
                                          min_value = 32, #no. of neurons will be 32 and 512
                                          max_value = 512,
                                          step = 32), #When sampling from it, the minimum step for walking through the interval is 32.
                           activation = 'relu')) #activation fucntion in the hidden layer 
  model.add(layers.Dense(1,activation = 'linear')) #output layer with one neuron since it is a regression problem 
  model.compile(
      optimizer=keras.optimizers.Adam(
            hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4])), #choice means out of the 3 we will select one 
        loss='mean_absolute_error',
        metrics=['mean_absolute_error']
  )
  return model
  
```

```python
tuner  = RandomSearch(
    build_model,
    objective = 'val_mean_absolute_error',
    max_trials = 5,
    executions_per_trial = 3,
    directory = 'project',
    project_name = 'price_prediction')
```

```python
tuner.search_space_summary()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

The fit method is calculating the mean and variance of each of the features present in our data. The transform method is transforming all the features using the respective mean and variance.

Using the transform method we can use the same mean and variance as it is calculated from our training data to transform our test data. Thus, the parameters learned by our model using the training data will help us to transform our test data.

fit_transform() is used on the training data so that we can scale the training data and also learn the scaling parameters of that data.


```python
#Scaling down the datas 
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_test
```

```python
#Training the model and checking for best hyperparameters 
tuner.search(X_train, Y_train,
             epochs=5,
             validation_data=(X_test, Y_test))
```
